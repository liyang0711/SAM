```python
from __future__ import division
```

```python
from netCDF4 import Dataset
from sklearn.decomposition import pca
import pandas as pd
from netcdftime import num2date
import numpy as np
from numpy import ma 
from scipy.stats import zscore
from matplotlib.mlab import detrend_linear
```

```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

## local file on disk 

```python
nc = Dataset('/home/nicolasf/data/NCEP/hgt.mon.mean.nc', 'r')
```

```python
nc.variables.keys()
```

```python
level = nc.variables['level'][:]
```

## will select 850 

```python
sel_level = 850 
ilevel = np.where(level == sel_level)
```

## get the lat and lon coordinates 

```python
lat = nc.variables['lat'][:]
lon = nc.variables['lon'][:]
```

## get the time

```python
origin = nc.variables['time'].units
time = num2date(nc.variables['time'][:], units=origin)
```

## select latitudes <= 20S 

```python
ilat = np.where(lat <= -20.)[0]
```

```python
hgt = nc.variables['hgt'][:,ilevel,ilat,:]
lat = lat[ilat]
```

## Calculates a matrix of weights

```python
weight = np.sqrt(np.cos(np.deg2rad(lat)))
```

```python
weight[-1] = 10e-9
```

```python
weight = np.repeat(weight[...,np.newaxis],len(lon),axis=1)
weight = np.repeat(weight[np.newaxis,...],hgt.shape[0],axis=0)
```

```python
hgt = hgt.reshape(hgt.shape[0], hgt.shape[1] * hgt.shape[2])
weight = weight.reshape(weight.shape[0], weight.shape[1] * weight.shape[2])
```

## casts the HGT array into a pandas DataFrame for slicing and dicing

```python
hgt_df = pd.DataFrame(hgt, index=time)
```

## calculates the climatology over 1981 - 2010

```python
hgt_clim = hgt_df.loc['1981':'2010',:]
```

```python
clim = hgt_clim.groupby(hgt_clim.index.month).mean()
```

## calculates the anomalies

```python
anoms = hgt_df - np.resize(clim, hgt_df.values.shape)
```

## standardizes

```python
X = zscore(anoms.values)
```

## multiply by the weights

```python
X = X * weight
```

## performs EOF

```python
skpca = pca.PCA().fit(X)
```

```python
var_explained = skpca.explained_variance_ratio_[0:10]
```

### shows variance explained 

```python
var_explained
```

### gets the Principal components 

```python
PCs = skpca.transform(X)
```

### gets the EOFs

```python
EOFs = skpca.components_
```

### reshapes and plots (just to check)

```python
EOFs = np.array([EOFs[i,:].reshape((len(lat), len(lon))) for i in xrange(EOFs.shape[0])]) # reshape each EOF
```

im = plt.imshow(EOFs[0,:,:], origin='upper', interpolation='nearest', aspect='auto'); plt.colorbar(im)

### calculates correlations and plots

```python
EOF1r = [np.corrcoef(PCs[:,0], X[:,i])[0][1] for i in xrange(X.shape[1])]
```

```python
EOF1r = np.array(EOF1r).reshape(len(lat), len(lon))
```

```python
from mpl_toolkits.basemap import Basemap as bm
from mpl_toolkits.basemap import addcyclic
```

```python
EOF1, lons = addcyclic(EOF1r, lon)
```

```python
f, ax = plt.subplots(figsize=(8,8))
m = bm(projection='spstere',lon_0=0.,boundinglat=-20.,resolution='l', round=True, ax=ax)
lons, lats = np.meshgrid(lons, lat)
cf = m.contourf(lons,lats,EOF1,np.arange(-0.8, 0.9, 0.1), cmap=plt.get_cmap('RdBu_r'), latlon=True, extend='both')
m.drawcoastlines(color='0.8', linewidth=2)
m.drawmeridians(np.arange(-180., 180., 40), labels=[1,1,1,1])
m.drawparallels(np.arange(-90., -20., 20), labels=[1,1,1,1])
m.colorbar(cf, pad='10%', label="Pearson's R")
```

### standardizes the first PC

```python
SAM = zscore(PCs[:,0])
```

### detrends the first PC

```python
SAM_detrend = detrend_linear(SAM)
```

### casts that into a pandas DataFrame

```python
SAM  = pd.DataFrame({'SAM':SAM, 'SAM_detrend':SAM_detrend}, index=hgt_df.index)
```

### selects the period of interest

```python
SAM = SAM.loc['1970':'2011']
```

### plots

```python
f, ax = plt.subplots(figsize=(10,6))
SAM.plot(ax=ax)
```
